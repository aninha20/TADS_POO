# Exercício 03

<p style="text-align:justify;">
Faça uma classe em Java que fique em laço solicitando a digitação de números inteiros maiores ou iguais a zero. Quando o usuário digitar um número menor que zero, o programa deve exibir a quantidade de números digitados e a média desses números.

</p>

```java
import java.util.Scanner;

//public class Main {
//  public static void main(String[] args){
        int quant = 0;
        int soma = 0;
        int numero = 0;
        
        Scanner scanner = new Scanner(System.in);
    
        while(numero >= 0){
            System.out.println("Digite um número inteiro, caso queira encerrar o programa digite um número menor que zero");
            numero = scanner.nextInt();
            
            if(numero >= 0){
                soma = soma + numero;
                quant++;
            }
        }
        
        if(quant > 0){
            System.out.printf("Quantidade de números digitados: %d\n", quant);
            System.out.printf("Média dos números digitados: %d\n", soma/quant);
        }else{
            System.out.println("Nenhum número válido foi digitado");
        }
        
//    }
//}
```
