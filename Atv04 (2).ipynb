# Exercício 01
<p style="text-align:justify;">
Crie um programa que armazene um vetor com as notas de dez alunos, calcule e imprima a média dessas notas e depois informe a quantidade de notas acima e abaixo da média calculada.
</p>

```java
import java.util.ArrayList;
import java.util.List;

//public class Main{
//    public static void main(String[] args){
        double vetorNotas[] = {10, 9, 4.5, 5, 7.8, 3.3, 8.9, 7, 3, 10};
        List<Double> acimaMedia = new ArrayList<Double>();
        List<Double> abaixoMedia = new ArrayList<Double>();
        double media, soma = 0;

        for(double nota : vetorNotas){
            soma += nota;
        }

        media = soma/vetorNotas.length;

        for(double nota : vetorNotas){
            if(nota > media){
                acimaMedia.add(nota);
            }else{
                abaixoMedia.add(nota);
            }
        }

        System.out.println("As notas que estão acima da média: " + acimaMedia);
        System.out.println("As notas que estão abaixo da média: " + abaixoMedia);
//    }
//}
```

<p style="text-align:justify;">
Construa uma classe em Java que seja capaz de efetuar a leitura de uma matriz de valores reais de tamanho 3 x 3, calculando seu determinante.
</p>

```java
import java.util.Scanner;

//public class DeterminanteMatriz{
//    public static void main(String[] args){
        double determinante = 0;

        Scanner scanner = new Scanner(System.in);

        // Leitura da matriz 3x3
        double matriz[][] = new double[3][3];
        System.out.println("Digite os elementos da matriz 3x3:");

        for(int i = 0; i < 3; i++){
            for(int j = 0; j < 3; j++){
                System.out.print("Digite o elemento da posição [" + i + "][" + j + "]: ");
                matriz[i][j] = scanner.nextDouble();
            }
        }

        // Cálculo do determinante seguindo a regra de Sarrus
        determinante += matriz[0][0] * (matriz[1][1] * matriz[2][2] - matriz[1][2] * matriz[2][1]);
        determinante -= matriz[0][1] * (matriz[1][0] * matriz[2][2] - matriz[1][2] * matriz[2][0]);
        determinante += matriz[0][2] * (matriz[1][0] * matriz[2][1] - matriz[1][1] * matriz[2][0]);

        System.out.println("O determinante da matriz é: " + determinante);

        scanner.close();
//    }
//}
```

Sua classe deve conter um laço de repetição simulando um menu que controle a entrada e a saída dos veículos com as opções: 1-Entrada, 2-Saída e 3-Listar situação atual e 4-Encerrar o programa.
```java
//decidi seprarar as ações de entrada, saída e listagem em métodos para melhor organização e entendimento
import java.util.Scanner;

//public class Estacionamento{
//   public static void main(String[] args){
        Scanner scanner = new Scanner(System.in);
        String[] vagas = new String[10];
        int opcao;

        do {
            System.out.println("\nMenu:");
            System.out.println("1- Entrada");
            System.out.println("2- Saída");
            System.out.println("3- Listar situação atual");
            System.out.println("4- Encerrar o programa");
            System.out.print("Escolha uma opção: ");
            opcao = scanner.nextInt();

            switch(opcao){
                case 1:
                    entradaVeiculo(vagas, scanner);
                    break;
                case 2:
                    saidaVeiculo(vagas, scanner);
                    break;
                case 3:
                    listarSituacaoAtual(vagas);
                    break;
                case 4:
                    System.out.println("Encerrando o programa...");
                    break;
                default:
                    System.out.println("Opção inválida. Tente novamente.");
            }
        } while(opcao != 4);

        scanner.close();
  }

    public static void entradaVeiculo(String[] vagas, Scanner scanner){
        System.out.print("Digite o número da vaga: ");
        int numeroVaga = scanner.nextInt();

        if (numeroVaga < 0 || numeroVaga >= vagas.length){
            System.out.println("Número de vaga inválido.");
            return;
        }

        System.out.print("Digite a placa do veículo: ");
        String placa = scanner.next();

        vagas[numeroVaga] = placa;
        System.out.println("Veículo estacionado com sucesso na vaga " + numeroVaga);
    }

    public static void saidaVeiculo(String[] vagas, Scanner scanner){
        System.out.print("Digite o número da vaga que será liberada: ");
        int numeroVaga = scanner.nextInt();

        if(numeroVaga < 0 || numeroVaga >= vagas.length){
            System.out.println("Número de vaga inválido.");
            return;
        }

        if(vagas[numeroVaga] == null){
            System.out.println("Não há veículo estacionado nesta vaga.");
            return;
        }

        vagas[numeroVaga] = null;
        System.out.println("Vaga liberada com sucesso.");
    }

    public static void listarSituacaoAtual(String[] vagas){
        System.out.println("Situação atual do estacionamento:");

        for(int i = 0; i < vagas.length; i++){
            if(vagas[i] != null){
                System.out.println("Vaga " + i + ": " + vagas[i]);
            } else {
                System.out.println("Vaga " + i + ": vazia");
            }
        }
//   }
//}
//Professor, por algum motivo o código só funciona aqui quando eu comento as duas últimas chaves, não entendi, pois, uma delas fecha o método.
```
